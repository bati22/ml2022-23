#### Zwolnienie z egzaminu
Osoby z najlepszymi projektami (ocena 5.0 z projektu) oraz osoby z oceną 5.0 z ćwiczeń będą zwolnione z egzaminu i automatycznie otrzymają ocenę 5.0 z całego przedmiotu.

```
# Ocena z ćwiczeń
punktacja_z_cwiczen = punktacja_z_notebookow + punktacja_z_projektu
if ocena_z_projektu == 5:
    ocena_z_cwiczen = 5
else:
    ocena_z_cwiczen = skala_uj(punktacja_z_cwiczen)

# Ocena z przedmiotu
if ocena_z_cwiczen == 5:
    ocena_z_przedmiotu = 5
else:
   punktacja = (punktacja_z_egzaminu + punktacja_z_cwiczen) / 2
   ocena_z_przedmiotu = skala_uj(punktacja)

# Egzamin
if ocena_z_przedmiotu == 5:
    return zwolnienie_z_egzaminu
elif ocena_z_przedmiotu >= 3.0:
    return pierwszy_termin_egzaminu
else:
    return drugi_termin_egzaminu
```

### Ocena końcowa

Studenci, którzy otrzymali pozytywną ocenę z ćwiczeń (co najmniej 3.0) mają prawo i obowiązek przystąpić do egzaminu pisemnego. Egzamin końcowy w pierwszym terminie przeprowadzony będzie w formie pisemnej. Ocena końcowa z przedmiotu jest oceną łączną z ćwiczeń i egzaminu. Za egzamin można otrzymać 0-100 pkt.

Punktacja końcowa = (punkty z egzaminu + punkty z ćwiczeń) / 2

Punkty końcowe przekładają się na ocenę końcową w następujący sposób:
* 0-49 pkt: 2
* 50-59 pkt: 3
* 60-69 pkt: 3,5
* 70-79 pkt: 4
* 80-89 pkt: 4,5
* 90-100 pkt: 5

W wypadku uzyskania w pierwszym terminie oceny niedostatecznej, student ma prawo podejść do egzaminu w DRUGIM TERMINIE.
